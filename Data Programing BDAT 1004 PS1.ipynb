        Console.WriteLine(" ");

        Console.WriteLine(" ");

    static double CalculateTriangleArea(double a, double b, double c)
    {
        // Calculate the semi-perimeter
        double s = (a + b + c) / 2.0;

        // Calculate the area using Heron's formula
        double area = Math.Sqrt(s * (s - a) * (s - b) * (s - c));

        // Return the calculated area
        return area;
    }
}

You can turn a word into pig-Latin using the following two rules(simplified):
• If the word starts with a consonant, move that letter to the end and append 'ay'. For example, 'happy' becomes 'appyhay' and 'pencil' becomes 'encilpay'.
• If the word starts with a vowel, simply append 'way' to the end of the word.
For example, 'enter' becomes 'enterway' and 'other' becomes 'otherway'. For our purposes, there are 5 vowels: a, e, i, o, u (so we count y as a consonant).
Write a function pig() that takes a word (i.e., a string) as input and returns its pigLatin form. Your function should still work if the input word contains upper case characters. Your output should always be lower case however.

```python
# Answer
# The program uses pig() function to accept any word as input and returns its pig- Latin form.
# Reference link1: https://www.datacamp.com/tutorial/case-conversion-python
# Reference Link2: https://stackoverflow.com/questions/57150426/what-is-printf

def pig(word):
    vowels = 'aeiouAEIOU'  # Define a string of vowels
    word = word.lower()  # Convert the input word to lowercase for consistent processing
    first_letter = word[0]  # Get the first letter of the lowercase word

    # Check if the first letter is in the set of vowels
    if first_letter in vowels:
        return word + 'way'  # If it's a vowel, add 'way' to the end of the word
    else:
        return word[1:] + first_letter + 'ay'  # If it's a consonant, move the first letter to the end and add 'ay'

# Accept user input
print("The program uses pig() function to accept any word as input and returns its pig- Latin form.")
print(" ")
print(" ")
user_input = input("Please enter any word: ")
print(" ")
# Call the pig() function with the user's input and store the result
result = pig(user_input)

# Display the result with a message
print(" ")
print(f"Pig('{result}')")
print(" ")
print(" ")
print(" ")
print(" ")
```

# Question 7 Python
File bloodtype1.txt records blood-types of patients (A, B, AB, O or OO) at a clinic.
Write a function bldcount() that reads the file with name name and reports (i.e.,
prints) how many patients there are in each bloodtype.

```python
# Answer
# The program uses the function bldcount() to read a file with name bloodtype1.txt and reports how many patients there are in each bloodtype
# Reference Link1: https://stackoverflow.com/questions/19034822/unknown-python-expression-filename-r-path-to-file
# Reference Link2: https://swcarpentry.github.io/python-novice-inflammation/09-errors.html

def bldcount(file_path):
    # Initialize a dictionary to store the count of each blood type
    blood_count = {'A': 0, 'B': 0, 'O': 0, 'AB': 0, 'OO': 0}

    try:
        # Open and read the file
        with open(file_path, 'r') as file:
            for line in file:
                # Remove whitespace and split the line into characters
                characters = line.strip()

                # Create a list to group "AB" and "OO" together
                grouped_characters = []
                i = 0

                # Process each character in the line
                while i < len(characters):
                    # Check if "AB" or "OO" is present
                    if i < len(characters) - 1 and (characters[i] + characters[i + 1]).upper() in ['AB', 'OO']:
                        grouped_characters.append((characters[i] + characters[i + 1]).upper())
                        i += 2
                    else:
                        grouped_characters.append(characters[i])
                        i += 1

                # Count the occurrences of blood types (A, B, O, AB, OO)
                for char in grouped_characters:
                    if char in blood_count:
                        blood_count[char] += 1

        # Print the counts for each blood type
        print("The program uses the function bldcount() to read a file with name bloodtype1.txt and reports how many patients there are in each bloodtype")
        print(" ")
        for blood_type, count in blood_count.items():
            if count == 1:
                print(f"There is 1 patient with blood type {blood_type}") 
                print(" ")
            elif count == 0:
                print(f"There are no patients with blood type {blood_type}")
                print(" ")
                print(" ")
                print(" ")
                print(" ")
            else:
                print(f"There are {count} patients with blood type {blood_type}")
                print(" ")
                
    except FileNotFoundError:
        print("File not found.")
    except Exception as e:
        print(f"An error occurred: {e}")

# Specify the file path
file_path = r'C:\Users\emeka\OneDrive\Desktop\Data Programming\Problem Set 1\bloodtype1.txt'

# Call the bldcount function with the file path
bldcount(file_path)
```

# Question 10 Python
Encryption is the process of hiding the meaning of a text by substituting letters in the message with other letters, according to some system. If the process is successful, no
one but the intended recipient can understand the encrypted message. Cryptanalysis refers to attempts to undo the encryption, even if some details of the encryption are
unknown (for example, if an encrypted message has been intercepted). The first step of cryptanalysis is often to build up a table of letter frequencies in the encrypted text.
Assume that the string letters is already defined as 'abcdefghijklmnopqrstuvwxyz'. 
Write a function called frequencies()that takes a string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text. Your function may ignore any characters that are not in letters

```python
# Answer
#The program uses the frequencies() function to accept string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text.
# Reference Link1: https://www.geeksforgeeks.org/python-frequency-of-each-character-in-string/
# Reference Link2: https://stackoverflow.com/questions/40985203/counting-letter-frequency-in-a-string-python

def frequencies(text):
    # Define the set of valid characters to count
    letters = 'abcdefghijklmnopqrstuvwxyz'
    
    # Initialize a list to store the frequency of each character
    char_frequency = [0] * 26
    
    # Convert the input text to lowercase to make it case-insensitive
    text = text.lower()
    
    for char in text:
        if char in letters:
            # Calculate the index for the character in 'letters' and increment the count
            index = letters.index(char)
            char_frequency[index] += 1
    
    return char_frequency

# Accept user input
print("The program uses the frequencies() function to accept string as its only parameter, and returns a list of integers, showing the number of times each character appears in the text.")
print(" ")
user_input = input("Please enter any text to count character frequencies: ")
result = frequencies(user_input)

# Display the character frequencies
for char, count in zip('abcdefghijklmnopqrstuvwxyz', result):
    print(f"{char}: {count}")
```
